```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop_duplicates(inplace=True)

df['director'] = df['director'].fillna("Not Available")
df['cast'] = df['cast'].fillna("Not Available")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Not Rated")
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['year_added'] = df['date_added'].dt.year

df[['duration_int', 'duration_type']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
df['duration_int'] = pd.to_numeric(df['duration_int'], errors='coerce')

df['year_added'] = df['year_added'].fillna(-1).astype(int)  
df['date_added'] = df['date_added'].fillna(pd.NaT)

df['duration_int'] = df['duration_int'].fillna(0).astype(int)
df['duration_type'] = df['duration_type'].fillna("Unknown")
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="type", palette="Set2")
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

```python
#Movies constitute a larger share than TV Shows, but TV content is substantial and strategic for engagement.
```

```python
plt.figure(figsize=(10,5))
df['year_added'].value_counts().sort_index().plot(kind="bar", color="teal")
plt.title("Content Added to Netflix Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
#A sharp ramp-up after ~2015, peaking around 2018–2020. Slight tapering afterwards likely reflects catalogue curation and shifting strategy.
```

```python
plt.figure(figsize=(8,4))
df['country'].value_counts().head(10).plot(kind="bar", color="orange")
plt.title("Top 10 Content-Producing Countries on Netflix")
plt.ylabel("Number of Titles")
plt.show()
```

```python
#The USA dominates, with India and UK often near the top. “Unknown” entries reflect incomplete metadata for some titles.
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, y="rating", order=df['rating'].value_counts().index, palette="coolwarm")
plt.title("Distribution of Ratings on Netflix")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

```python
#TV-MA (mature) leads, suggesting a strong focus on adult-oriented content; family/teen categories are present but smaller shares.
```

```python
movie_durations = df[df['type'] == "Movie"].dropna(subset=["duration_int"])
plt.figure(figsize=(8,5))
sns.histplot(movie_durations['duration_int'], bins=30, kde=True, color="purple")
plt.title("Distribution of Movie Durations on Netflix")
plt.xlabel("Duration (minutes)")
plt.ylabel("Count")
plt.show()
```

```python
#Most movies cluster around 80–120 minutes; the mean typically lands near ~100 minutes, aligning with theatrical norms.
```

```python
tv_show_seasons = df[df['type'] == "TV Show"].dropna(subset=["duration_int"])
plt.figure(figsize=(8,5))
sns.countplot(data=tv_show_seasons, x="duration_int", palette="Set1")
plt.title("Distribution of TV Show Seasons on Netflix")
plt.xlabel("Number of Seasons")
plt.ylabel("Count")
plt.show()
```

```python
#Most shows have 1–3 seasons, indicating breadth of catalogue and limited multi-season runs relative to long-running series.
```

```python
tags = (df.assign(tag=df['listed_in'].str.split(','))
          .explode('tag'))
tags['tag'] = tags['tag'].str.strip()

top_tags = tags['tag'].value_counts().head(15)

plt.figure(figsize=(10,6))
top_tags.plot(kind='barh')
plt.title("Top Content Categories / Tags")
plt.xlabel("Number of Titles"); plt.ylabel("Tag")
plt.tight_layout(); plt.show()
```

```python
#“International TV Shows”, “Dramas”, “Comedies”, “Documentaries”, and “Stand-Up Comedy” frequently appear—diverse portfolio with strong 
#international presence.
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='rating', y='release_year', palette='Set3')
plt.title('Distribution of Release Year by Show Type')
plt.xticks(rotation=45)
plt.show()
```

```python
def remove_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return series[(series >= lower_bound) & (series <= upper_bound)]

# Filter DataFrame based on release_year
df_clean = df[df['release_year'].between(
    df['release_year'].quantile(0.25) - 1.5*(df['release_year'].quantile(0.75) - df['release_year'].quantile(0.25)),
    df['release_year'].quantile(0.75) + 1.5*(df['release_year'].quantile(0.75) - df['release_year'].quantile(0.25))
)]
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_clean, x='rating', y='release_year', palette='Set3')
plt.title('Distribution of Release Year by Show Type (Outliers Removed)')
plt.xticks(rotation=45)
plt.show()
```
